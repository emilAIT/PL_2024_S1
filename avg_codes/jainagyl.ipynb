```python
from csv import reader
import csv
from collections import defaultdict

clients_file = reader(open('clients.csv', 'r'))
next(clients_file)

distributors_file = reader(open('distributor.csv', 'r'))
next(distributors_file)
 
products_file = reader(open('products.csv', 'r'))
next(products_file)


clients = {idx: {'name': name, 'tel': tel, 'address': address.replace('\n', ' ')}for idx, name, tel, address in clients_file}
distributors = {idx: {'name': name, 'tel': tel}for idx, name, tel in distributors_file}
products ={idx: {'name':name, 'wieght': weight} for idx, name, weight in products_file}
purcahses = [i for i in reader(open('purchase.csv', 'r'))][1:]
sales = [i for i in reader(open('sales.csv', 'r'))][1:]


# 1. Напишите программу на Python, которая рассчитает общее количество каждого товара, купленного у дистрибьюторов, используя данные из файлов items.csv и purchase.csv.
def each_product_count():
    d = defaultdict(int)
    for _, _, item_id, count, _, _ in purcahses:
        d[products[item_id]['name']] += int(count)
    return d
# print(each_product_count())


# 2. Объедините файлы sales.csv и items.csv, чтобы рассчитать общий доход,  полученный от продажи каждого товара в магазине.
def merge_sales():
    d = defaultdict(float)
    for _, _, item_id, _, _, total_cost in sales:
        d[products[item_id]['name']] += float(total_cost)
    return d
# print(merge_sales())


#! 3. Используя данные из purchase.csv, distributor.csv и items.csv, рассчитайте общую сумму, потраченную каждым дистрибьютором на каждый товар, и определите, какой дистрибьютор поставил самый дорогой товар.
def most_income_tovar():
    d = defaultdict(float)
    for _, distributor_id, _, _, _, total_cost in purcahses:
        d[distributor_id] += float(total_cost)
    
    n = {}
    for key, val in d.items():
        if key == distributors[int(key)-1][0]:
            n[distributors[int(key)-1][1]] = val
    mx = 0 
    dst = None
    for _, distributor_id, _, price, _ in purcahses:
        if float(price) > mx:
            mx = float(price) 
            dst = distributor_id
    for distributor_id, distributor_name, _ in distributors:
        if distributor_id == dst:
            return distributor_name, mx 
    return n


# print(most_income_tovar())


# 4. Объедините sales.csv и clients.csv, чтобы выяснить, какой клиент совершил наибольшую сумму покупок в магазине.

def merge_clients():
    d = defaultdict(int)
    for _, client_id, _, _, _, total_cost in sales:
        d[clients[client_id]['name']] += float(total_cost)
    mx = 0
    mx_name = ''
    for name, count in d.items():
        if count > mx:
            mx = count
            mx_name = name
    return mx_name, mx

# print(merge_clients())


# 5. Используя файлы sales.csv, clients.csv и items.csv, найдите топ-5 клиентов по количеству разных товаров, которые они приобрели.

def top5():
    d = defaultdict(list)
    for _, client_id, item_id, _, _, _ in sales:
        if products[item_id] not in d[clients[client_id]['name']]:
            d[clients[client_id]['name']].append(products[item_id])
    tops = []
    for i in range(5):
        mx = 0 
        mx_name = 0
        for name, count in d.items():
            if len(count) > mx and name not in tops:
                mx = len(count)
                mx_name = name
        tops.append(mx_name)
    return tops
# print(top5())


#! 6. Объедините purchase.csv, items.csv и distributor.csv, чтобы рассчитать среднюю цену, заплаченную за каждый товар магазином у разных дистрибьюторов.
def average_every__for_product(): 
    d = defaultdict(list)
    for _, _, item_id, count, _, total_cost in purcahses:
        d[products[item_id]].append((count, total_cost))
    res = {}
    for name, value in d.items():
        sm = 0
        count = 0
        for count, total_cost in value:
            sm += float(total_cost)
            count += int(count)
        res[name] = sm / count
    return res


# print(average_every__for_product())



#7 # Напишите программу на Python, чтобы определить, какой товар приносит наибольшую прибыль, сравнив стоимость покупки (из purchase.csv) и цену продажи (из sales.csv).


def most_profit():
    d = defaultdict(lambda: {'sale': 0, 'purchase': float('+inf')})
    for  _, item_id, _, price_per_item , in sales:
        if float(price_per_item) > float(d[item_id]['sale']):
            d[item_id]['sale'] = float(price_per_item)
    for _, _,item_id, _,price_per_item, _ in purcahses:
        if float(price_per_item) < float(d[item_id]['purchase']):
            d[item_id]['purchase'] = float(price_per_item)
    mx = 0
    mx_name = ''
    for k, v in d.items():
        income = v['sale'] - v['purchase']
        if income > mx:
            mx = income 
            mx_name = k
    return products[mx_name]['name'], mx 
# print(most_profit())


# 8. Объедините purchase.csv и sales.csv, чтобы проанализировать, продал ли магазин больше товаров, чем приобрел, за определенный период времени.



def more_than(date1, date2):
    count_sale, count_purchase = 0, 0
    for date, _, _,count, _, _ in sales:
        if date1 <= date <= date2:
            count_sale += int(float(count))
    for date, _, _,count, _, _ in purcahses:
        if date1 <= date <= date2:
            count_purchase += int(count)
    return count_sale, count_purchase, count_sale - count_purchase  > 0
# print(more_than('2023-07-16', '2023-07-26'))



# 9. Используя clients.csv, sales.csv и items.csv, напишите программу,  чтобы определить клиента, который купил наибольшее количество определенного товара.

def get_client(product):
    d = defaultdict(int)
    for  _,client_id,item_id,count, _, _ in sales:
        if products[item_id]['name'] == product:
            d[clients[client_id]['name']] += int(float(count))
    mx, mx_name = 0, ''
    for name, count in d.items():
        if count > mx:
            mx = count
            mx_name = name
    return mx_name, mx 

# print(get_client('three'))

    

#10. Напишите программу на Python, которая использует данные из distributor.csv и purchase.csv, чтобы узнать, какой дистрибьютор поставил больше всего разных товаров в магазин.

def most_distributor():
    d = defaultdict(list)
    for  _,distributor_id, item_id, _, _, _ in purcahses:
        if item_id not in d[distributors[distributor_id]['name']]:
            d[distributors[distributor_id]['name']].append(item_id)
    mx, mx_name = 0, ''
    for name, arr in d.items():
        if len(arr) > mx :
            mx = len(arr)
            mx_name = name 
    return mx_name, mx 
# print(most_distributor())


#11. Объедините sales.csv и items.csv, чтобы рассчитать среднюю цену продажи каждого товара в магазине.
def all_average():
    d = defaultdict(lambda: {'total': 0, 'count': 0})
    for  _, _,item_id,count, _,total_cost in sales:
        d[products[item_id]['name']]['total'] +=float(total_cost)
        d[products[item_id]['name']]['count'] += float(count)
    text = '\n'
    for name, value in d.items():
        avg = int(value['total'] / value['count'])
        text += f'{name} --> {avg}\n'
    return text
# print(all_average())  



#! 12. Напишите программу на Python, чтобы рассчитать общий доход от клиентов, которые купили товары у определенного дистрибьютора, используя данные из sales.csv, clients.csv и distributor.csv.



#13. Используя items.csv, purchase.csv и sales.csv, выясните, какой товар имеет наибольший товарооборот (т.е. наиболее часто покупаемый и продаваемый)

def most_product():
    d = defaultdict(lambda: {'sale': 0, 'purchase': 0 })
    for  _, _,item_id,count, _, _ in purcahses:
        d[products[item_id]['name']]['purchase'] += int(count)
    for  _, _,item_id,count, _, _ in sales:
        d[products[item_id]['name']]['sale'] += int(float(count))
    mx_sale, mx_purchase, mx_name = 0, 0, ''
    for name, values in d.items():
        if values['sale'] > mx_sale and values['purchase'] > mx_purchase:
            mx_sale = values['sale']
            mx_purchase = values['purchase']
            mx_name = name
    return mx_name
# print(most_product())


#14. Объедините purchase.csv, distributor.csv и sales.csv, чтобы узнать, какой доход от продаж был получен от товаров, поставляемых каждым дистрибьютором.
def most_profit_persent():
    d1 = defaultdict(list)
    for  _,distributor_id,item_id, _, _, _ in purcahses:
        if item_id not in d1[distributors[distributor_id]['name']]:
            d1[distributors[distributor_id]['name']].append(item_id)
    total = 0
    for _ ,_ ,_ ,_ ,_ ,total_cost in sales:
        total += float(total_cost)
    d2 = defaultdict(float)
    for _ ,_ ,item_id,_ ,_ ,total_cost in sales:
        d2[item_id] += float(total_cost) / total * 100
    text = '\n'
    diors = sorted(list(d1.keys()))
    for idx, name in enumerate(diors):
        persent = 0
        for i in d1[name]:
            persent += d2[i]
        text += f'{idx + 1}) {name} --> {int(persent)}%\n'
    return text
# print(most_profit_persent())
```
